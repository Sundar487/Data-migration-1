```python
import logging
import boto3
from botocore.exceptions import ClientError


def create_bucket(bucket_name, region="us-east-2"):
    """Create an S3 bucket in a specified region

    If a region is not specified, the bucket is created in the S3 default
    region (us-east-1).

    :param bucket_name: Bucket to create
    :param region: String region to create bucket in, e.g., 'us-west-2'
    :return: True if bucket created, else False
    """

    # Create bucket
    try:
        if region is None:
            s3_client = boto3.client('s3')
            s3_client.create_bucket(Bucket=bucket_name)
        else:
            s3_client = boto3.client('s3', region_name=region)
            location = {'LocationConstraint': region}
            s3_client.create_bucket(Bucket=bucket_name,
                                    CreateBucketConfiguration=location)
    except ClientError as e:
        logging.error(e)
        return False
    return True
```

```python
import json
import os

bucket_name = 'aws-json-from-website'
folder_name = 'jsonfile/'

db_name = "awsdb"  # database name
coll_name = "submission"  # collection name

# Get objects for the database and the collection
db = client[db_name]
coll = db[coll_name]

# List all objects in the S3 folder
result = s3.list_objects(Bucket=bucket_name, Prefix=folder_name)

# Loop through each object in the S3 folder
for content in result.get('Contents', []):
    # Check if the object is a JSON file
    if content['Key'].endswith('.json'):
        # Get the object from S3 and load the JSON data
        obj = s3.get_object(Bucket=bucket_name, Key=content['Key'])
        data = json.load(obj['Body'])
        
        # Set the document ID to the filename without the extension
        file_base_name = os.path.basename(content['Key'])
        data['_id'] = os.path.splitext(file_base_name)[0]
        
        # Insert the document into the collection
        coll.insert_one(data)
        
print('success')
```
